Rank of a tensor

```python
rank1_tensor = tf.Variable(["Nancy", "One"], tf.string)
rank2_tensor = tf.Variable([['test', 'ok'], ['Nancy', 'one']], tf.string)
```

```python
tf.rank(rank1_tensor)
```

```python
tf.rank(rank2_tensor)
```
